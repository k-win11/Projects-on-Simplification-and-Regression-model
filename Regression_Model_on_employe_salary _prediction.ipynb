# Import pakage from sklearn linearRegression and other one pandas

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

# Catogories data in traget Y and feature X

```python
df=pd.read_csv("D:DUCAT/DS_AI/dataset/regression/salary_1_variable.csv")

X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values



```

# fit the model 


```python
model=LinearRegression()
model.fit(X,y)
```

Using the matplotlib for graphical representation also checking the out liers 

```python
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
plt.scatter(df.YearsExperience,df.Salary,color = 'y')
plt.show()
```

```python
df['pred']=model.predict(X)
```

```python
plt.scatter(df.YearsExperience,df.pred,color='g')

plt.show()
```

```python
plt.scatter(df.YearsExperience,df.Salary,color = 'y')
plt.scatter(df.YearsExperience,df.pred,color='g')
plt.show()
```

```python
from sklearn.metrics import accuracy_score

model.score(X,y)
```
